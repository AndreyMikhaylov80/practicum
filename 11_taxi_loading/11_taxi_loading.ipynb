### Функция для создания признаков
Напишем функцию для создания признаков, а именно:
* день;
* день недели;
* час;
* отстающие значения;
* скользящее среднее:

```python
def make_features(data, max_lag, rolling_mean_size):
    data['day'] = data.index.day
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour

    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    data = data.dropna()
```

### Создадим признаки
Применим функцию с лагом до 23 часов и скользящим средним равным 15. Удалим пропуски и посмотрим данные:

```python
make_features(taxi, 23, 15)
taxi = taxi.dropna()
taxi.head()
```

### Разделение данных для обучения
Поделим данные в соотношении 1:9 и выделим таргет

```python
train, test = train_test_split(taxi, shuffle=False, test_size=0.1)

features_train = train.drop(['num_orders'], axis=1)
target_train = train['num_orders']
features_test = test.drop(['num_orders'], axis=1)
target_test = test['num_orders']

print('Размер обучающих выборок:', features_train.shape, target_train.shape)
print('Размер тестовых выборок:', features_test.shape, target_test.shape)
```

### Функция для расчета метрики RMSE:

```python
def RMSE(target, predict):
    return (mean_squared_error(target, predict))**0.5

#С помощию функции make_scorer создаю метрику RMSE
RMSE_score = make_scorer(RMSE, greater_is_better=False)
```

### Модели

Так как мы работаем с временными рядами, то будем использовать метод Time series split, который обеспечивает репрезентативность обучающих и тестовых наборов данных. Чтобы тестовый набор не содержал информации, которая была недоступна в момент создания обучающего набора.

#### Линейная регрессия

```python
%%time
# Разделим тренировочные  данные на 5 частей для перекрестной проверки временных рядов
tscv = TimeSeriesSplit(n_splits=5).split(features_train)

# объявляем модель линейной регрессии
regression = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=-1, normalize=True)
# normalize=True - нормализовать входные переменные
# fit_intercept=True - вычислять отрезок b₀ в уравнении типа f(x) = b₀ + b₁x
# n_jobs= -1 (все доступные) - количество процессов, задействованных в параллельных вычислениях.
# copy_X=True - копировать входные переменные

# Обучаем модель
regression.fit(features_train, target_train)

# Определяем RMSE модели с помощью кросс-валидации
regression_cv_score = pd.Series(cross_val_score(regression, features_train, target_train, scoring=RMSE_score, cv=tscv)).mean()*(-1)
print('Среднее качество модели линейной регрессии на кросс-валидации:', regression_cv_score)
```

#### Регрессия случайного леса

```python
%%time
# Разделим тренировочные  данные на 5 частей для перекрестной проверки временных рядов
tscv = TimeSeriesSplit(n_splits=5).split(features_train)

# Объявляем модель случайной лесной регрессии:
forest = RandomForestRegressor()
# Определяем словарь параметров:
forest_parametrs = { 'n_estimators': range(5, 30, 5),
                     'max_depth': range(1, 15, 2),
                     'min_samples_leaf': range(1, 5),
                     'min_samples_split': range(2, 8, 2) }

# Применим GridSearchCV с кросс-валидацией:
forest_grid = GridSearchCV(forest, forest_parametrs, scoring=RMSE_score, cv=tscv)
forest_grid.fit(features_train, target_train)
```

Посчитаем качество модели при оптимальных гиперпараметрах:

```python
%%time

forest_params = forest_grid.best_params_
forest_score = forest_grid.score(features_train, target_train) * (-1)
print(forest_params)
print(forest_score)
print( )
```

#### Модель CatBoost

```python
%%time

# Разделим тренировочные данные на 5 частей для перекрестной проверки временных рядов
tscv = TimeSeriesSplit(n_splits=5).split(features_train)

# Объявлям модель градиентного бустинга CatBoost:
cb = cb.CatBoostRegressor()

# Определим словарь параметров:
cb_parametrs = {'learning_rate': [0.03, 0.1],
                'depth': [4, 6, 10]
               }

# Применим GridSearchCV с кросс-валидацией:
cb_grid = GridSearchCV(cb, cb_parametrs, scoring=RMSE_score, cv=tscv)
cb_grid.fit(features_train, target_train, verbose=500)
```

Посчитаем качество модели при оптимальных гиперпараметрах:

```python
%%time

cb_params = cb_grid.best_params_
cb_score = cb_grid.score(features_train, target_train) * (-1)
print(cb_params)
print(cb_score)
print()
```

### Подготовка функции для построения графика важности признаков
Объявляю функцию для построения графика важности признаков, она будет показывать только те признаки, важность которых больше 0.0:

```python
def feature_importance_hist(model, features_train):
    importances = list(model.feature_importances_)
    names_indices = features_train.columns
    not_null_importances = []
    not_null_indices = []

    for i in importances:
        if i>0:
            not_null_importances.append(i)
            not_null_indices.append(names_indices[importances.index(i)])

    plt.figure(figsize=(17, 6))
    plt.title("Feature importances")

    plt.bar(range(len(not_null_importances)), not_null_importances, color="g")
    plt.xticks(range(len(importances)), not_null_indices, rotation=90)

    plt.tight_layout()
    plt.xlim([-1, len(not_null_importances)])
    plt.show()
```
